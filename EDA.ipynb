```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dm_office_sales.csv')
df.head(10)
```

```python
df.info()
```

```python
# determining whether our columns are categorical or continuous(numeric)
```

```python
#categorical
df['division'].value_counts()
```

```python
#categorical
df['level of education'].value_counts()
```

```python
#categorical
df['training level'].value_counts()
```

```python
#categorical
df['work experience'].value_counts()
```

```python
#continuous
df['salary'].value_counts()
```

```python
#continuous
df['sales'].value_counts()
```

```python
df.describe().transpose()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x= df['salary'], y= df['sales'], hue= 'level of education');
```

```python
# visualizzing the sales, salary and work experience
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x= df['salary'], y= df['sales'],hue='work experience');
```

```python
# visualizing sales, salary and training level. to see whether there aare coorelations
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x= df['salary'], y= df['sales'],size='training level')
```

```python
# we choose countplot because it will help us check the relationship between categoricaal and numerical data set
# thus comparing categorical dataa and continous data
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='division', hue='level of education') # want compare the individual effects on the division column
```

```python
# finding the distribution of salary
plt.figure(figsize=(10,8))
sns.displot(data=df, x='salary')
```

```python
plt.figure(figsize=(10,6))
sns.rugplot(data=df,x='salary', hue='level of education', height= 0.8)
```

```python
# alternatively we can use kernel density plot. thus tells you the probability of finding sth.
plt.figure(figsize=(10,6))
sns.kdeplot(data=df,x='salary', hue='level of education')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='salary')
```

```python
sns.jointplot(data=df, x='salary', y='sales', kind='scatter')
```
